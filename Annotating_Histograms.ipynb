```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://docs.google.com/spreadsheets/d/1VMaw2oCn0ABitd-alLAEsEhGS1Je2UFNLu76TKrIH7w/gviz/tq?tqx=out:csv&sheet=Raw_Medical_Data_for_day1'
df = pd.read_csv(url, index_col = 0)
df.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x='Income', ax = ax);
```

```python
# Find the mean and round to 2 decimal places for readability
mean = df['Income'].mean().round(2)
mean
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x = 'Income', ax = ax)
ax.axvline(mean);
```

```python
# Change line color to red
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x='Income', ax = ax)
ax.axvline(mean, color = 'red');
```

```python
# Add legend and title
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x = 'Income' ,ax=ax)
ax.axvline(mean, color = 'red', label='Mean')
ax.set_title('Income')
ax.legend();
```

```python
# Add legend with the value of the mean
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x='Income', ax=ax)
ax.axvline(mean, color='red', label=f'Mean {mean}');
ax.set_title('Income')
ax.legend();
```

```python
def plot_annotated_histogram(df, col):
    ## Saving the mean of the column
    mean = df[col].mean().round(2)
    # Plotting the histogram + new col's mean
    fig,ax = plt.subplots(figsize=(10,6))
    sns.histplot(data=df, x=col, ax=ax)
    ax.axvline(mean, color='red',label=f'Mean {mean}');
    ax.set_title(col)
    ax.legend();
    return fig, ax
```

```python
fig,ax = plot_annotated_histogram(df, col = 'Initial_days')
```

```python
## Updating our function to add the median
def plot_annotated_histogram(df, col):
    ## Saving the mean of the column
    mean = round(df[col].mean(), 2)
    median = round(df[col].median(), 2)
    ## Creating the Plot
    fig,ax = plt.subplots(figsize=(10,6))
    sns.histplot(data=df, x=col, ax=ax)
    ax.set_title(col)
    ## Annotating measures of centrality
    ax.axvline(mean, color='red',label=f'Mean: {mean}', lw=2);
    ax.axvline(median, color = 'green',  label = f'Median: {median}', ls="--", lw=2)

    ax.legend();
    return fig, ax
```

```python
fig, ax = plot_annotated_histogram(df, col='Initial_days')
```
