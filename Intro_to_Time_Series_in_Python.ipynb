A `datetime` object like `datetime.datetime(2026, 6, 7, 21, 59, 14, 463941)` represents a specific point in time and contains several components:

*   **2026**: This is the **year**.
*   **6**: This is the **month** (June).
*   **7**: This is the **day** of the month.
*   **21**: This is the **hour** (21:00 is 9 PM).
*   **59**: This is the **minute**.
*   **14**: This is the **second**.
*   **463941**: This is the **microsecond** (one millionth of a second).

```python
# Assuming 'now' is your datetime object, you can access each part using its attributes:

print(f"Year: {now.year}")
print(f"Month: {now.month}")
print(f"Day: {now.day}")
print(f"Hour: {now.hour}")
print(f"Minute: {now.minute}")
print(f"Second: {now.second}")
print(f"Microsecond: {now.microsecond}")
```

#**Datetime.strftime**

##For the day of the month:

```python
# just output the day with a leading 0 if needed
format_a = "%d"
now.strftime(format_a)
```

# For month:

```python
# Add full month name to day of month
format_b = "%B %d"
now.strftime(format_b)
```

# For the day of the week:

```python
# Add day of the week to month and day
format_c = "%a, %B %d"
now.strftime(format_c)
```

# For the year:

```python
# include abbreviated month and 2 digit year
format_d = "%b %y"
now.strftime(format_d)
```

# For Time:

```python
## the default datetime format
default ="%Y-%m-%d %H:%M:%S.%f"
print(now)
now.strftime(default)
```

```python
# Example US Formatted 12-hour date/time with AM/PM
fmt = "%m/%d/%Y %I:%M:%S %p"
now.strftime(fmt)
```

```python
## just the time (24 hour time)
now.strftime("%T")
```
