```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# This filepath works if you have added the AXSOS ACADEMY file structure to your Drive
fpath="/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/ames-housing-cleaned-eda.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
# What categories are in Bldg Type?
df['Bldg Type'].value_counts()
```

# **Bar Plots with Matplotlib Only**

```python
# Use groupby to calculate the average for each category
bldg_type_prices = df.groupby('Bldg Type')['SalePrice'].mean()
bldg_type_prices
```

```python
# Plot the average of each as a bar
fig, ax = plt.subplots()
ax.bar(bldg_type_prices.index, bldg_type_prices.values);
ax.set_ylabel('SalePrice')
ax.set_xlabel("Bldg Type");
```

#**Bar Graph Using Pandas**

```python
# Using pandas to plot the means
ax = bldg_type_prices.plot(kind='bar')
ax.set_ylabel("Sale Price");
```

#**Bar Graph Using Seaborn**

```python
# Using seaborn to plot the group means
ax = sns.barplot(data=df, x='Bldg Type', y="SalePrice")
```

#**Creating Grouped Bar Charts (using Seaborn's hue argument)**

```python
# Using seaborn to plot the group means by Buildign Type colored by Central Air
ax = sns.barplot(data=df, x="Bldg Type", y="SalePrice", hue='Central Air')
```

```python
# Using seaborn to plot the group means by central air colored by Bldg Type
ax = sns.barplot(data=df, x="Central Air", y="SalePrice", hue='Bldg Type')
```

#**Correlation in Python**

```python
# Get correlations
corr = df.corr(numeric_only = True)
corr.round(2)
```

```python
# Make a heatmap of correlations
sns.heatmap(corr, cmap = 'Greens');
# Include annotations on the heatmap
sns.heatmap(corr, cmap = 'Greens', annot = True);
```

```python
# Create a fig, ax with a figsize argument
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(corr, cmap = 'coolwarm', annot = True, ax=ax);
```

#**Regression Plots**

```python
# We can make a scatter plot with pandas' .plot.
ax = df.plot(kind='scatter', x='Bedroom', y='SalePrice');
```

```python
# Scatter Plot with Matplotlib
fig, ax = plt.subplots()
ax.scatter(df['Bedroom'], df['SalePrice']);
ax.set(xlabel='Bedroom', ylabel="SalePrice");
```

```python
sns.regplot(data=df, x="Bedroom", y="SalePrice")
```

```python
# Saving the arguments for the line color in a dict to to use as line_kws
line_kws = dict(color='black')
ax = sns.regplot(data=df, x="Bedroom", y="SalePrice",
                 line_kws=line_kws)
```

```python
# Saving the arguments for the markers' edge color in a dict to to use as scatter_kws
scatter_kws = dict(edgecolor='white')
ax = sns.regplot(data=df, x="Bedroom", y="SalePrice",
                 line_kws=line_kws, scatter_kws=scatter_kws);
```

```python
# Creating a scatterplot using Central Air as the hue argument
ax = sns.scatterplot(data=df, x="Bedroom" , y="SalePrice",
                     hue="Central Air");
```

```python
# Use sns.lmplot and save as g
g = sns.lmplot(data=df, x="Bedroom", y="SalePrice",
                hue="Central Air");
```

```python
# Adding back the scatter_kws
g= sns.lmplot(data=df, x="Bedroom", y="SalePrice",
               scatter_kws=scatter_kws,
                hue="Central Air");
```

```python
# Set the title using g.ax.set_title()
g = sns.lmplot(data=df, x="Bedroom", y="SalePrice",
               scatter_kws=scatter_kws,
                hue="Central Air");
g.ax.set_title("LM Plot");
```
