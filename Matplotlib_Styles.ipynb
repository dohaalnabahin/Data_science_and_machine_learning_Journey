```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Keep in mind that the filename will be different on your computer
fpath = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week03/kc_house_data.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
# Previewing the first 30 entries in rcParams
list(plt.rcParams.keys())[:30]
```

```python
# Checking the default figure size
plt.rcParams['figure.figsize']
```

```python
# Changing the default figure size
plt.rcParams['figure.figsize'] = (7,5)
plt.rcParams['figure.figsize']
```

```python
# Updating many params at once
plt.rcParams.update({
        "font.weight": "bold",  # bold fonts
        "lines.linewidth": 1,   # thick lines
        "lines.color": "k",     # black lines
        "grid.color": "0.5",    # gray gridlines
        "grid.linestyle": "-",  # solid gridlines
        "grid.linewidth": 0.5,  # thin gridlines
        "savefig.dpi": 300,     # higher resolution output.
        })
```

```python
# To see every style available run the following:
plt.style.available
```

```python
# temporarily testing the fivethirtyeight style
with plt.style.context('fivethirtyeight'):
    # any plots created in this block will use the style
    ax = df['price'].hist(bins = 30, edgecolor='black')
    ax.tick_params(axis='x', rotation = 45)
    ax.ticklabel_format(style='plain')
    ax.set_title('Style = fivethirtyeight')
```

```python
# temporarily testing the ggplot style
with plt.style.context('ggplot'):
    # any plots created in this block will use the style
    ax = df['price'].hist(bins = 30, edgecolor='black')
    ax.tick_params(axis='x', rotation = 45)
    ax.ticklabel_format(style='plain')
```

```python
with plt.style.context(['ggplot','fivethirtyeight']):
    # any plots created in this block will use the style
    ax = df['price'].hist(bins = 30, edgecolor='black')
    ax.tick_params(axis='x', rotation = 45)
    ax.ticklabel_format(style='plain')
    ax.set_title('Style = ggplot + fivethirtyeight')
```

```python
# temporarily testing the ['fivethirtyeight','ggplot'] combined style
with plt.style.context(['fivethirtyeight','ggplot']):
    # any plots created in this block will use the style
    ax = df['price'].hist(bins = 30, edgecolor='black')
    ax.tick_params(axis='x', rotation = 45)
    ax.ticklabel_format(style='plain')
    ax.set_title('Style = fivethirtyeight + ggplot')
```

```python
# Setting the style for the notebook
plt.style.use(['fivethirtyeight','ggplot'])
```

```python
# New plot will automatically use the style
ax = df['price'].hist(bins = 30, edgecolor='black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

```python
## For every available style
for style_name in plt.style.available:
    # Temporarily use the style
    with plt.style.context(style_name):
        # Create the graph
        ax = df['price'].hist(bins = 30, edgecolor='black')
        ax.tick_params(axis='x', rotation = 45)
        ax.ticklabel_format(style='plain')
        # Change the graph title to the style name
        ax.set_title(style_name)
        # To see each style separately, we must show the graph in our loop
        plt.show()
```

```python
def my_scatterplot(df, x='bathrooms', y='sqft_living'):
    fig, ax = plt.subplots()
    ax.scatter(df[x], df[y], edgecolor='white')
    ax.set( title=f'{x} vs. {y}', ylabel=y, xlabel=x);
    return fig, ax
```

```python
fig, ax = my_scatterplot(df)
```

```python
context_name = 'poster'
with sns.plotting_context(context_name):
    fig,ax = my_scatterplot(df)
    ax.set_title(context_name)
```

```python
## temporarily changing context
for context_name in ['notebook','paper','poster','talk']:
    with sns.plotting_context(context_name):
        fig,ax =  my_scatterplot(df)
        ax.set_title(context_name)
        plt.show()
```

```python
with sns.plotting_context("talk", font_scale=0.9):
    fig,ax =  my_scatterplot(df)
    ax.set_title('Context="talk", font_scale=0.9')
```

```python
## making the settings permanent for the notebook:
sns.set_context("talk", font_scale=0.9)
fig,ax =  my_scatterplot(df)
```
