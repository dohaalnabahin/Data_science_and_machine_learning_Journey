# Class Attributes

الصفات (Attributes):

هي الخصائص التي تهمنا في التطبيق (مثل: الاسم، الإيميل، رصيد الحساب).

الدالة السحرية __init__:

هي دالة "تجهيز" الكائن. بمجرد أن تكتبي User()، يقوم بايثون تلقائياً بتشغيل هذه الدالة ليعطي الكائن الجديد صفاته.

الاستدعاء:

هذه الدالة تُنفذ فقط مرة واحدة لكل كائن، وهي لحظة "ولادته" (Instantiation)

Screenshot (1147).png

```python
class User:   # (Class Definition)

    # 1. حذفنا name و email من هنا لأنك وضعت قيم ثابتة بالداخل
    def __init__(self):

        # 2. تعيين الصفات (Attributes)
        self.name = "Michael"
        self.email = "michael@axsos-academy.com"
        self.account_balance = 0

# --- هنا نبدأ من أول السطر (خارج الكلاس) ---

# 3. إنشاء الكائنات (الأقواس فارغة الآن لتناسب تعريف الدالة فوق)
guido = User()
monty = User()

# 4. الوصول للبيانات (Accessing Attributes)
print(guido.name)   # سيطبع: Michael
print(monty.email)  # سيطبع: michael@axsos-academy.com
```

# عملية بناء الكلاس (User)
## 1. تعريف الكلاس والصفات (Attributes)

نبدأ بتعريف القالب الأساسي. الصفات هي المعلومات التي يمتلكها الكائن، ونضعها داخل الدالة السحرية __init__. ولاحظي أننا نستخدم self دائماً لربط الصفة بالكائن

```python
class User:
    def __init__(self):
        # تعريف صفات افتراضية لكل مستخدم جديد
        self.name = "Michael"
        self.email = "michael@axsos-academy.com"
        self.account_balance = 0
```

## 2. إنشاء النسخ (Instantiating)

هنا نقوم بصنع "كائنات" حقيقية من القالب. بما أن الدالة __init__ لا تطلب مدخلات غير self حالياً، نترك الأقواس فارغة


```python
guido = User() # إنشاء الكائن الأول
monty = User() # إنشاء الكائن الثاني
```

## 3. الوصول إلى الصفات (Accessing Attributes)

بمجرد إنشاء الكائن، يمكننا "قراءة" بياناته باستخدام اسم الكائن متبوعاً بنقطة ثم اسم الصفة.

```python
print(monty.name) # المخرجات: Michael
print(monty.email) # المخرجات: Michael
print(monty.account_balance) # المخرجات: Michael
```

 ## 4. تعديل الصفات (Changing Attribute Values)

أهم ميزة هي أننا نستطيع تغيير البيانات لكل كائن بشكل منفصل بعد إنشائه، بحيث لا يظلان متشابهين.



```python
guido.name = "Doha" # غيرنا اسم الكائن الأول فقط
monty.name = "Samir" # غيرنا اسم الكائن الثاني فقط

print(guido.name) # المخرجات: Guido
print(monty.name) # المخرجات: Monty
```


 #تمرير البيانات عند الإنشاء (Passing Arguments)

## المعاملات (Parameters):

نحدد في دالة __init__
البيانات التي يجب توفرها (مثل الاسم والإيميل).

## القيم الافتراضية (Default Values):

 الصفات التي نعرف أن قيمتها واحدة للجميع
 (مثل الرصيد $0) لا نطلبها كمعامل، بل نثبتها بالداخل.

## الربط (Mapping):

 نقوم بربط القيمة القادمة من الخارج
 (مثلاً username) بالصفة الداخلية للكائن (self.name).

```python
class User2:
    # نطلب اسم المستخدم وإيميله فقط كشرط للإنشاء
    def __init__(self, username, email_address):
        self.name = username            # تخزين الاسم الممرر
        self.email = email_address      # تخزين الإيميل الممرر
        self.account_balance = 0        # قيمة ثابتة تبدأ من صفر للجميع
```

```python
# الآن نمرر البيانات الحقيقية بين الأقواس
guido = User2("Guido van Rossum", "guido@python.com")
monty = User2("Monty Python", "monty@python.com")

# عند الطباعة، كل كائن سيطبع بياناته الخاصة التي أرسلناها له
print(guido.email)   # المخرج: Guido van Rossum
print(monty.name)   # المخرج: Monty Python
```

# Class Methods
ما هي الـ Methods؟

هي ببساطة دوال (Functions)
لكنها تعيش داخل الكلاس. الفرق الجوهري هو أنها لا تعمل من تلقاء نفسها، بل يجب أن "يأمرها" كائن معين
(Instance) لكي تنفذ المهمة


## أهم مميزات الـ Methods:


مرتبطة بالكائن:

عندما يقوم "مايكل" بإيداع مبلغ، يجب أن يزيد رصيد "مايكل" فقط،
 وليس رصيد "آنا". الـ
 Method تعرف مَن الذي يناديها عن طريق كلمة self.

طريقة الاستدعاء:

 لا نناديها باسمها مباشرة مثل deposit()
 ، بل يجب كتابة اسم الكائن أولاً:
  guido.deposit()

```python
class User3:
    # 1. المُنشئ: يحدد صفات المستخدم عند ولادة الكائن
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.account_balance = 0  # الرصيد يبدأ دائماً من 0

    # 2. ميثود الإيداع: تزيد رصيد المستخدم بمبلغ معين
    def make_deposit(self, amount):
        # self.account_balance تشير لرصيد الكائن الذي استدعى الدالة
        self.account_balance += amount

# --- خارج الكلاس: تجربة الكود ---

# إنشاء مستخدم جديد
guido = User3("Guido van Rossum", "guido@python.com")
monty = User3("monty van Rossum", "monty@python.com")

# تنفيذ عملية إيداع
guido.make_deposit(100) # أودعنا 100
guido.make_deposit(200) # أودعنا 200 أخرى
monty.make_deposit(50) # أودعنا 50 أخرى



# التأكد من الرصيد
print(guido.name)             # المخرج: Guido van Rossum
print(guido.account_balance)  # المخرج: 300

print(monty.name)             # المخرج: Guido van Rossum
print(monty.account_balance)  # المخرج: 300

```

# Notes

لأن بايثون بيعتبر الكائن اللي قبل النقطة هو المعامل الأول.

لما تكتبي: guido.make_deposit(100)

المعلومة الأولى (الهوية):

هي guido (موجودة قبل النقطة)،
 بايثون بياخدها
  وبيدخلها لـ self تلقائياً.

المعلومة الثانية (البيانات):

 هي 100 (موجودة داخل الأقواس)،
  وبتروح لـ amount.

النتيجة: أنتِ بتكتبي 1 داخل الأقواس، وبايثون بيسحب 1 من قبل النقطة، المجموع صار 2.. وهيك الميثود اكتملت عندها الشروط!

وضحت ليش بنكتب بس وحدة (أو ثنتين) في الأقواس؟ لأن الكائن نفسه هو "المعامل المخفي"

لو نادى Guido الدالة
