```python
# `plt` is alwyas used as the alias for the `matplotlib.pyplot` module
import matplotlib.pyplot as plt
```

```python
# seaborn is always imported with the alias "sns"
import seaborn as sns
```

```python
import pandas as pd
# Loading csv data directly from URL
url="https://docs.google.com/spreadsheets/d/e/2PACX-1vQPk0YEO46WmL9yZSuO-2lTnTrfFz9keW4ittWKP0RkRpNwrpJkK9tlM11AIui4RSEeJlP9YlLtLOJS/pub?output=csv"
df = pd.read_csv(url)
df.head()
```

```python
# إنشاء الإطار والرسمة
fig, ax = plt.subplots()
```

```python
# تحديد بيانات المحور الأفقي (X) والمحور الرأسي (y1)
X = df['Time (hours)']
y1 = df['Runner 1 Distance']
```

```python
# إنشاء الإطار والرسمة
fig, ax = plt.subplots()

# رسم الخط
ax.plot(X, y1);
```

#**Line plot with ax.plot()**


```python
# You can also plot another line on the same graph
y2 = df['Runner 2 Distance']
fig, ax = plt.subplots()
ax.plot(X, y1)
ax.plot(X, y2);
```

#**Scatter Plot with ax.scatter()**

```python
# Use ax.scatter to plot the values without a line
fig, ax = plt.subplots()
ax.scatter(X, y1)
ax.scatter(X, y2);
```

#**Bar Charts with ax.bar()**

```python
# حساب المتوسط لكل عداء
runner1_mean = y1.mean()
runner2_mean = y2.mean()

# تحديد الأسماء التي ستظهر تحت كل عمود
bar_labels = ["Runner 1", "Runner 2"]
```

```python
# إنشاء الإطار والرسمة
fig, ax = plt.subplots()

# رسم الأعمدة
ax.bar(bar_labels, [runner1_mean, runner2_mean]);
```

#**Adding Labels and a Legend**

```python
fig, ax = plt.subplots()

# رسم الخط الأول مع إضافة تسمية (label)
ax.plot(X, y1, label = "Runner 1")

# رسم الخط الثاني مع إضافة تسمية (label)
ax.plot(X, y2, label = "Runner 2")

# إظهار مفتاح الرسم الذي يحتوي على التسميات
ax.legend();
```

#**Adding Axis Labels and Title**

## Separate lines of code

```python
# إنشاء الإطار والرسمة ورسم الخطوط
fig, ax = plt.subplots()
ax.plot(X, y1, label = "Runner 1")
ax.plot(X, y2, label = "Runner 2")
ax.legend()

# 1. إضافة تسمية للمحور الرأسي (Y)
ax.set_ylabel("Distance (miles)")

# 2. إضافة تسمية للمحور الأفقي (X)
ax.set_xlabel("Time (hours)")

# 3. إضافة العنوان الرئيسي للرسمة
ax.set_title("Comparing Runner Distances Over Time");

```

## A single line of code with ax.set()

```python
# Adding labels for each line and a legend
fig, ax = plt.subplots()
ax.plot(X, y1, label = "Runner 1")
ax.plot(X, y2, label = "Runner 2")
ax.legend()
# Using ax.set to customize
ax.set(ylabel= "Distance (miles)", xlabel = "Time (hours)", title = "Comparing Runner Distances Over Time");
```
