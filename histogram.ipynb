```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',100)
```

```python
fpath="/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week03/Data/ames-housing-cleaned-eda.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
# Make a default histogram of the price column
df['SalePrice'].hist();
```

```python
ax = df['SalePrice'].hist()
```

```python
# Remove scientific notation
ax = df['SalePrice'].hist()
ax.ticklabel_format(style='plain');
```

```python
# rotate your xticklabels
ax = df['SalePrice'].hist()
ax.tick_params(axis='x', rotation=45)
ax.ticklabel_format(style='plain');
```

```python
# Change number of bins to 30
ax = df['SalePrice'].hist(bins = 30)
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

```python
# Change number of bins to auto
ax = df['SalePrice'].hist(bins = 'auto')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

```python
# Change number of bins t0 30 and add edgecolor
ax = df['SalePrice'].hist(bins = 30, edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

```python
# Add title and labels
ax = df['SalePrice'].hist(bins = 30, edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
ax.set_title('Distribution of Home Prices')
ax.set_xlabel('Price ($)')
ax.set_ylabel('Number of Houses');

```

```python
# Use seaborn's histplot function and save the ax
ax = sns.histplot(data=df, x="SalePrice")
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x', rotation = 45);
```

```python
# Use binwidth arg to make histogram with $50k bins
ax = sns.histplot(data=df, x='SalePrice', binwidth=50_000)
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x', rotation = 45);
```

```python
# Use binwidth arg to make histogram with $50k bins and plot as % of total
ax = sns.histplot(data=df, x='SalePrice', binwidth=50_000, stat='percent')
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x', rotation = 45);
```

**Our histogram displays the distribution of home prices in the dataset. The majority of homes are priced at less than $500,000. A small percentage of homes cost $0 - $50,000 (approximately 1-2%), a larger number of homes were priced between $50,000 - $100,000 (around 13%), but the most common price range is between $100,000 - $150,000 (approximately 37%).**
