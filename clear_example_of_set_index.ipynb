```python
import pandas as pd
# import numpy
import numpy as np

names_list = ['brownie','cookie','cake', 'cupcake']
prices_list = [2.25,1.25,9.5, 3.5]
quantities_sold_list = [17, 40, 1, 10]

prices_array = np.array(prices_list)
quantities_sold_array = np.array(quantities_sold_list)

prices_array * quantities_sold_array

df = pd.DataFrame({'Name':names_list,
                   'Price':prices_list,
                   "Quantity Sold": quantities_sold_list})


df.columns

df.values

df.index # يخزن معادلة بدل اؤقام


 #   يعرض فقط الجدول انو غير الاندكس من ارقام لاسماء الاعمدة يعني لو رجعنا طبعنا الجدول الاصلي ما هغير الاندكس من ارقام لاسماء الاعمدة
df.set_index("Name")

 # هنا هيتم التغيير على الجدول الاصلي يعني لو طبعا الجدول الاصلي هيطبع بدل الاندكس اسماء الاعمدة
df = df.set_index("Name")

# هنا غير الناتج من معادلة لاسماء الاعمدة صارو هما الاندكس
df.index


```

```python
# saving sales as a new column
df['Sales'] = df['Price'] * df["Quantity Sold"]
df
```

```python
# Get the sum of the Sales column
df['Sales'].sum()
```

```python
# Making a series from a dictionary
expenses = pd.Series({'brownie':0.25, 'cookie':0.5, 'cake':5, 'cupcake': 0.75},
                     name='Expenses')
expenses
```

```python
# saving our combined data as df
df = pd.concat([df, expenses], axis=1)
df
```

```python
# Calculate profit and define it as new column
df['Profit Per Item'] = df['Price'] - df['Expenses']
df
```

```python
# Calculate the profit per type of baked good and define as a column
df["Profits"] = df['Profit Per Item'] * df['Quantity Sold']
df
```

```python
# Find total profit
total_profit = df['Profits'].sum()
total_profit
```

```python
# What percentage of the goal has been reached?
total_profit/1000
```
