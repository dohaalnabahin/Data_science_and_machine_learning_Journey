# Fitness Tracker



```python
class Fitness_Track:
    def __init__(self, user_name, goal=10000):
        self.user_name = user_name
        self.goal = goal
        self.steps = 0
        self.calories_burned = 0

    def log_steps(self, amount):
        self.steps += amount
        self.calories_burned += (amount / 20)
        return self

    def reset_day(self):
        self.steps = 0
        self.calories_burned = 0
        return self

    def status(self):
        print(f"--- {self.user_name}'s Progress ---")
        print(f"Steps: {self.steps} | Calories: {self.calories_burned}")

        if self.steps >= self.goal:
            print("Goal Achieved! You are a Hero! üèÜ")
        else:
            x = self.goal - self.steps
            print(f"Keep moving! You need {x} more steps to reach your goal.")
        return self


user1 = Fitness_Track("Lana", 3000)
user1.log_steps(100).log_steps(2500).status()
```
