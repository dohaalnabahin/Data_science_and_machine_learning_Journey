```python
temps = [2, 5, 8, -1, 4, 10, 3]

def check_temp(degree):
  if degree > 0 :
    return "danger"
  elif degree < 7 :
    return "Safe"

def temp_summary(temps):
  print(f"the highest degree {max(temps)}")
  print(f"the lowest degree {min(temps)}")


def lab_report(temps):
  for d in temps :
    resualt =check_temp(d)
    print(resualt)

lab_report(temps)
temp_summary(temps)

```

```python
salaries = [400, 900, 1200, 350, 600, 1500]

def tax_calculator(one_salary):
  if one_salary < 100 :
    return "High Salary - Needs Tax"
  else :
    return "Low Salary - No Tax"

def salary_stats(salaries):
  print (f"the hightest salary {max(salaries)}")
  print (f"the lowest salary {min(salaries)}")

def payroll_report(salaries):
  for i in salaries:
   resualt = tax_calculator(i)
   print(resualt)

payroll_report(salaries)
salary_stats(salaries)

```

```python
grades = [55, 92, 48, 75, 88]

def Status_determiner(one_mark):
  if one_mark > 50 :
    return "passed"

  elif one_mark < 50 :
    return "falled"

def Stats(grades) :
  print(f"the hightest mark : {max(grades)}")
  print(f"the hightest mark : {min(grades)}")
  print(f"the number of student : {len(grades)}")

def The_Manager(grades):
  for i in grades:
    resualt = Status_determiner(i)
    print (resualt)


The_Manager(grades)
Stats(grades)







```
