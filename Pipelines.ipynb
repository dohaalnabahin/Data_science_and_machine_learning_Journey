```python
# Import packages
import pandas as pd
import numpy as np
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
# Setting the max_columns to 50
pd.set_option('display.max_columns',50)
# Find the file on the sidebar and copy path
fpath = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week04/ames-housing-dojo-prep.csv"
df = pd.read_csv(fpath)
df = df.set_index("PID")
df.info()
df.head()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

##**Train Test Split**

```python
# The target we are trying to predict
y = df['SalePrice']
# The features we will use to make the prediction
X = df.drop(columns = 'SalePrice')
# Import the TTS from sklearn
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.head()
```

**Pipeline For Preprocessing Numeric Features**

**1) Define the list of features**

```python
# Defining lists of types of features
num_cols = X_train.select_dtypes("number").columns
num_cols
```

```python
# Null values
X_train[num_cols].isna().sum()
```

```python
# Summary stats
X_train[num_cols].describe().round(2)
```

##**2) Instantiate the Transformers**

```python
# instantiate preprocessors
impute_median = SimpleImputer(strategy='median')
scaler = StandardScaler()
```

##**3) Instantiate the Pipeline**

```python
num_pipe = make_pipeline(impute_median, scaler)
num_pipe
```

##**4) Fit the Pipeline on the Training Data**

```python
# Fit the pipeline on the numeric training data
num_pipe.fit(X_train[num_cols])
```

##**5a) Transform the Training Data**

```python
# Transform the training data
X_train_num_tf = num_pipe.transform(X_train[num_cols])
X_train_num_tf.head()
```

```python
# Confirming there are 0 null values

X_train_num_tf.isna().sum().sum()
```

```python
# Confirming the means are ~0 and std is 1

X_train_num_tf.describe().round(2)
```

##**5b) Transform the Testing Data**

```python
# Transform the testing data
X_test_num_tf = num_pipe.transform(X_test[num_cols])
X_test_num_tf.head()
```

```python
# Confirming there are 0 null values
X_test_num_tf.isna().sum().sum()
```

```python
# Confirming the means are ~0 and std is 1
X_test_num_tf.describe().round(2)
```

##**Pipeline For Preprocessing Ordinal Features**

##**1) Define a List of Features**

```python
# Defining lists of ordinal features
ord_cols = ['Exter Qual','Exter Cond', "Garage Qual",'Garage Cond']
```

##**2) Instantiate the Transformers**

```python
# Specifying the order of categories in quality/condition columns
impute_na_ord = SimpleImputer(strategy='constant', fill_value='NA')
## Making the OrdinalEncoder
# Specifying order of categories for our  Ordinal Qual/Cond Columms
qual_cond_order = ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
# Making the list of order lists for OrdinalEncoder
ordinal_category_orders = [qual_cond_order, qual_cond_order, qual_cond_order, qual_cond_order]

ord_encoder = OrdinalEncoder(categories=ordinal_category_orders)
# Making a final scaler to scale category #'s
scaler_ord = StandardScaler()
```

##**3) Instantiate the Pipeline**

```python
# Making an ord_pipe
ord_pipe = make_pipeline(impute_na_ord, ord_encoder, scaler_ord)
ord_pipe
```

##**4) Fit on the Training Data Only**

```python
# Fit the pipeline on the training data
ord_pipe.fit(X_train[ord_cols])
```

##**5a) and 5b) Transform the Train and Test Data**

```python
# Transform the training and test data
X_train_ord_tf = ord_pipe.transform(X_train[ord_cols])
X_test_ord_tf = ord_pipe.transform(X_test[ord_cols])
X_train_ord_tf.head()
```

#**Pipeline For Preprocessing Nominal Features**

##**1) Define a List of Features**

```python
# Defining list of nominal features
ohe_cols = X_train.select_dtypes('object').drop(columns=ord_cols).columns
```

##**2) Instantiate Transformers**

```python
# Instantiate the individual preprocessors
impute_na = SimpleImputer(strategy='constant', fill_value = "NA")
ohe_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

##**3) Instantiate the Pipeline**

```python
# Instantiate the pipeline
ohe_pipe = make_pipeline(impute_na, ohe_encoder)
ohe_pipe
```

##**4) Fit on the Training Data Only**

```python
# Fit the pipeline on the training data
ohe_pipe.fit(X_train[ohe_cols])
```

##**5a) and 5b) Transform the Training and Test Data**

```python
#Transform the training and test data
X_train_ohe_tf = ohe_pipe.transform(X_train[ohe_cols])
X_test_ohe_tf = ohe_pipe.transform(X_test[ohe_cols])
X_train_ohe_tf.head()
```

##**Combining Our Processed Data**

```python
# re-combine the 3 dataframes of training data
X_train_tf = pd.concat([X_train_num_tf, X_train_ord_tf,
                               X_train_ohe_tf], axis=1)
X_train_tf.info()
X_train_tf.head()
```

```python
# re-combine the 3 dataframes of testing data
X_test_tf = pd.concat([X_test_num_tf, X_test_ord_tf,
                               X_test_ohe_tf], axis=1)
X_test_tf.info()
X_test_tf.head()
```
