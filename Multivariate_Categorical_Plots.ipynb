```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',100)
```

```python
# This filepath will work if you are using the recommended file structure
fname = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/ames-housing-cleaned-eda.csv"
df = pd.read_csv(fname)
df.head()
```

```python
# Checking how many categories in each categorical column
df.select_dtypes('object').nunique()
```

```python
# Univariate countplot
ax = sns.countplot(data=df, x='Bldg Type')
```

```python
# Multivariate countplot with a hue argument
ax = sns.countplot(data=df, x='Bldg Type', hue='Central Air')
```

```python
# Making a seaborn barplot as a starting point
ax = sns.barplot(data=df, x='Bldg Type', y='Lot Area')
```

```python
# Stripplot takes the same arguments as barplot
ax = sns.stripplot(data=df, x='Bldg Type', y='Lot Area',
                   hue='Bldg Type')
```

```python
ax = sns.barplot(data=df, x='Bldg Type', y='Lot Area',
                   hue='Bldg Type')
```

```python
# Customizing the stipplot
ax = sns.stripplot(data=df, x='Bldg Type', y='Lot Area',
                   hue='Bldg Type', edgecolor='white', linewidth=1)
```

```python
# Customizing the stipplot
ax = sns.stripplot(data=df, x='Bldg Type', y='Lot Area',
                   hue='Bldg Type', edgecolor='white', linewidth=1)
```

```python
 #Catplot with default kind is stripplot
g = sns.catplot(data=df, x='Bldg Type', y='Lot Area', hue='Bldg Type')
```

```python
 #Catplot with default kind is stripplot
g = sns.catplot(data=df, x='Bldg Type', y='Lot Area', hue='Bldg Type')
```

```python
# Using catplot to make a boxplot, adding dodge=False to reduce white space
g = sns.catplot(data=df, x='Bldg Type', y='Lot Area',hue='Bldg Type',
                kind='box', dodge=False)
```

```python
# Using catplot to make a violin plot
g = sns.catplot(data=df, x='Bldg Type', y='Lot Area',hue='Bldg Type',
                kind='violin', dodge=False)
```
