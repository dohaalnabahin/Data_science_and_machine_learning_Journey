# Logic Operators

1. قاعدة and (الشرطي الصارم)

تخيليها مثل "بوابة أمنية" تطلب تذكرتين؛ إذا نقصت واحدة، لن تفتح البوابة.

True + True ← النتيجة: True

True + False ← النتيجة: False

False + False ← النتيجة: False

example ⁉

(1 <= 2) and (2 <= 3) => True

(1 <= 2) and (2 >= 3) => False

(1 >= 2) and (2 >= 3) => False

تخيليها مثل "دعوة" تقبل أي شخص معه تذكرة واحدة على الأقل.

True + True ← النتيجة: True

False + False ← النتيجة: False (الوحيدة التي تعطي خطأ)

example ⁉

(1 <= 2) or (2 >= 3) => True

(1 <= 2) or (2 <= 3) => True

(1 >= 2) or (2 >= 3) => False

not True ← النتيجة: False

# Loops
### For Loops with Range

range()؟

هي دالة بتعطيكِ سلسلة أرقام. بدل ما تكتبي الأرقام واحد واحد، هي بتجهزهم إلك

ملخص المدخلات الثلاثة:

البداية (Start): من أي رقم يبدأ اللوب (شامل الرقم نفسه).

النهاية (Stop): أين يتوقف اللوب؟ (ملاحظة: لا يطبع هذا الرقم، يتوقف قبله بخطوة).

الزيادة (Step): كم مقدار القفزة في كل لفة؟ (افتراضياً هي 1).

الشكل الكامل (The Full Way):
range(0, 5, 1)  (ابدأ من 0، وقف عند 5، زيد 1).


اختصار الزيادة (Default Step):
range(0, 5)  بايثون بتفهم لحالها إنك بدك تزيد 1 في كل لفة.


الاختصار الأقصى (Default Start & Step):
range(5)  بايثون بتفهم إنك بدك تبدأ من 0 وتزيد 1



```python
# مدخل واحد فقط: range(stop)
#هنا بايثون يفترض أنكِ ستبدأين من 0 والزيادة هي 1
for i in range(5):
    print(i)
# الناتج: 0, 1, 2, 3, 4

# مدخلان: range(start, stop)
for i in range(2, 6):
    print(i)
# الناتج: 2, 3, 4, 5 (بدأ من 2 وتوقف قبل 6)

# ثلاثة مدخلات: range(start, stop, step)
for i in range(1, 10, 2):
    print(i)
# الناتج: 1, 3, 5, 7, 9 (بدأ من 1 وكل مرة زاد 2)

#*****************************************************************

# 1. الشكل الكامل: تحديد البداية، النهاية، والزيادة
for x in range(0, 10, 1):
    print(x)

# 2. اختصار الزيادة: بايثون سيفترض تلقائياً أنها +1
for x in range(0, 10):
    print(x)

# 3. الاختصار الأقصى: بايثون سيفترض البداية 0 والزيادة +1
for x in range(10):
    print(x)

# ملاحظة: كل هذه الحلقات (Loops) ستطبع الأرقام من 0 إلى 9
```

# Notes
لاحظ أنه إذا كنت بحاجة إلى تحديد زيادة غير +1، فإن جميع الوسائط الثلاثة مطلوبة

```python
# المثال الأول: القفز للأمام (زيادة بمقدار 2)
for x in range(0, 10, 2):
    print(x)
# المخرجات: 0, 2, 4, 6, 8

# ---------------------------------------

# المثال الثاني: العد التنازلي (نقصان بمقدار 3)
for x in range(5, 1, -3):
    print(x)
# المخرجات: 5, 2
```

بايثون توفر لكِ طريقتين للمرور على القائمة (List)، والفرق بينهم هو: هل يهمكِ "مكان" العنصر (Index) أم يهمكِ "العنصر نفسه"؟


الطريقة الأولى: باستخدام range و len (التركيز على الموقع)
نستخدمها إذا كنا نريد معرفة ترتيب العنصر (رقم الغرفة) بالإضافة للعنصر نفسه



```python
my_list = ["abc", 123, "xyz"]

#في البرمجة (وبايثون خصوصاً)، العد لا يبدأ من 1، بل يبدأ دائماً من 0

for i in range(0, len(my_list)):
    print(i, my_list[i])

# المخرجات:
# 0 abc
# 1 123
# 2 xyz
```

الطريقة الثانية: الدوران المباشر (التركيز على القيمة)
هذه الطريقة أسهل وأكثر استخداماً في بايثون ("Pythonic Way")، نستخدمها إذا كنا نريد القيم فقط

ليش بنحط  V
للوصول للقيمة: لكي نستطيع طباعة العنصر أو استخدامه، نحتاج لاسم يشير إليه.

الترتيب: بايثون تأخذ أول عنصر وتضعه في v وتنفذ الكود، ثم تأخذ الثاني وتضعه في v وتنفذ الكود، وهكذا.

```python
my_list = ["abc", 123, "xyz"]
for v in my_list:
    print(v)

# المخرجات:
# abc
# 123
# xyz
```

# 1. طباعة العناوين فقط (.keys())
نستخدمها إذا كنا نريد المفاتيح فقط

```python
capitals = {
      "palestine": "Gaza",
    "egypt": "cairo"
}
for key in capitals.keys():
    print(key)
# المخرجات: أسماء الدول/المدن فقط
```

# 2. طباعة المحتوى فقط (.values())
نستخدمها إذا كنا نريد القيم فقط.

```python
capitals = {
      "palestine": "Gaza",
    "egypt": "cairo"
}
for val in capitals.values():
    print(val)
# المخرجات: العواصم فقط
```

# 3. طباعة الاثنين معاً (.items())
وهذه "الأشمل"، تعطيكِ المفتاح والقيمة جنباً إلى جنب.

```python
capitals = {
      "palestine": "Gaza",
    "egypt": "cairo"
}
for key, val in capitals.items():
    print(key, val)
# المخرجات: الدولة = العاصمة
```

# While Loops
ي طريقة أخرى للتكرار طالما أن شرطًا معينًا صحيح

ملاحظة صغيرة: في الـ while loop ضروري جداً وجود السطر الأخير count += 1 لأنه هو الذي ينهي اللوب، وإلا سيستمر للأبد!

while<expression>
 do something, including progress towards making the expression False. Otherwise we'll never get out of here!


```python
count = 0
while count < 5:
    print("looping - ", count)
    count += 1
```

#  while with Else

تُستخدم لتنفيذ كود "أخير" يشتغل فقط إذا انتهى اللوب بشكل كامل وطبيعي (يعني خلص كل لفاته بدون ما ينقطع فجأة).

```python
y = 3
while y > 0:
    print(y)
    y -= 1
else:
    print("تم الانتهاء بنجاح!")
```

بإيجاز شديد، متى تعمل else مع اللوب؟

تشتغل: إذا انتهى اللوب "بسلام" (يعني خلص عدّ للأخير، أو إذا كان الشرط خطأ من البداية).

ما تشتغل: إذا قاطعنا اللوب بكلمة break أو return (يعني خروج مفاجئ).

الخلاصة: الـ else هي "مكافأة" اللوب الذي يكمل طريقه للنهاية دون انقطاع.

# break
"الهروب المفاجئ"؛ تُستخدم لإنهاء اللوب فوراً والخروج منه قبل أن يكمل دوراته، وذلك عند تحقق شرط معين (مثل العثور على ما تبحثين عنه).

نقاط سريعة:
في while و for: تشتغل مع الاثنين.

اللوبات المتداخلة: إذا وضعتِها داخل لوب وسط لوب، تخرجك من الداخلي فقط.

الهروب: بمجرد ما يشوفها بايثون، بيقفز للسطر اللي بعد اللوب مباشرة.

```python
for val in "string":
    if val == "i":
        break
    print(val)
```

 # continue
"زر التخطي"؛ تُستخدم لتجاهل (تطنيش) اللفة الحالية فقط، والقفز فوراً إلى اللفة التي بعدها دون الخروج من اللوب.

متى نستخدمها؟
عندما نريد استثناء عنصر معين (مثل رقم منحوس أو اسم محظور) ولكننا نريد إكمال باقي القائمة للنهاية

```python
for val in "string":
    if val == "i":
        continue
    print(val)
```
