```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Loading from URL
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSwI3TxQGHcIQWQ_Om7wPmSyN3t1sfOGYNbUn9i63tB1RVC0NFXKniaWy4hTB3XnsohGH9EXxfKTf7v/pub?output=csv"
df = pd.read_csv(url)
df.head()
```

```python
# Create filter for females and save as df_female
female_filter = df['Gender'] == "Female"
df_female = df.loc[female_filter]
df_female.head()
```

```python
# Save the opposite rows as df_male using ~female_filter
df_male = df.loc[~female_filter]
df_male.head()
```

```python
## making 2 side-by-side subplots
fig, axes = plt.subplots(ncols=2)
```

```python
## making 2 side-by-side subplots
fig, axes = plt.subplots(ncols=2, figsize=(10,5))
```

```python
## making 2 subplots on top of each other
fig, axes = plt.subplots(nrows=2, figsize=(6,8))
```

```python
## making 2 subplots side-by-side
fig, axes = plt.subplots(ncols=2, figsize=(10,5))
axes[0].set_title('axes[0]')
axes[1].set_title('axes[1]');
```

```python
# Create 2 side by side subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))
# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], color = 'blue')
axes[0].set_title('Males', fontsize = 12)
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")
# Subplot 2
axes[1].scatter(df_female['Height'], df_female['Weight'], color = 'orange')
axes[1].set_title('Females', fontsize = 12)
axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)");
```

```python
### SAME PLOTTING CODE AS ABOVE
# Create 2 side by side subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,4))
# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], c = 'blue')
axes[0].set_title('Males', fontsize = 12)
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")
# Subplot 2
axes[1].scatter(df_female['Height'], df_female['Weight'], c = 'orange')
axes[1].set_title('Females', fontsize = 12)
axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)")
### New: Adding a figure title (suptitle)
fig.suptitle("Comparing Height vs Weight")
```

```python
### SAME PLOTTING CODE AS ABOVE
# Create 2 side by side subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,4))
# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], c = 'blue')
axes[0].set_title('Males', fontsize = 12)
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")
# Subplot 2
axes[1].scatter(df_female['Height'], df_female['Weight'], c = 'orange')
axes[1].set_title('Females', fontsize = 12)
axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)")
# Adding a figure suptitle
fig.suptitle("Comparing Height vs Weight")
### New: fixing overlap with fig.tight_layout()
fig.tight_layout()
```

```python
# Adding sharey=True and sharex = True to use same axis limits for both plots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,4),
                         sharey=True, sharex=True);
# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], c = 'blue');
axes[0].set_title('Males', fontsize = 12);
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")
# Subplot 2
axes[1].scatter(df_female['Height'], df_female['Weight'], c = 'orange');
axes[1].set_title('Females', fontsize = 12);
### ylabel is no longer needed for the second subplot.
# axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)")
## Adding a figure title (superior title)
fig.suptitle("Comparing Height vs Weight (with sharex=True, sharey=True)",
             fontsize=14)
## Fixing overlapping labels
fig.tight_layout()
```

```python
### Creating the Figure and Axes with matplotlib
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (6,8), sharex=True)

# Plotting Subplot 1 with Seaborn
sns.scatterplot(data=df_male, x='Height', y='Weight', ax=axes[0])
axes[0].set_title('Male Height vs Weight', fontsize = 12);
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")

# Plotting Subplot 2 with Pandas
df_male['Height'].hist(bins='auto', edgecolor='black',ax=axes[1])
axes[1].set_ylabel('Counts per Bin', fontsize = 12);
axes[1].set_xlabel("Height (cm)")
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12,8))
# Adding titles to the top row
axes[0,0].set_title("axes[0,0]", fontweight="bold")
axes[0,1].set_title("axes[0,1]", fontweight="bold")
axes[0,2].set_title("axes[0,2]", fontweight="bold")
## Adding titles to the bottom row
axes[1,0].set_title("axes[1,0]", fontweight="bold")
axes[1,1].set_title("axes[1,1]", fontweight="bold");
axes[1,2].set_title("axes[1,2]", fontweight="bold");
fig.suptitle("Comparing Indices for 2D Subplots",
             fontweight='bold', fontsize=14)
fig.tight_layout()
```

```python
## Creating subplots with 2 rows and 2 columns, sharing the x-axis
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8),
                         sharex=True)
## Plotting male data
# Subplot 0,0 (top-left)
sns.scatterplot(data=df_male, x='Height', y='Weight', ax=axes[0,0])
axes[0,0].set_title('Male Height vs Weight', fontsize = 12);
axes[0,0].set_ylabel("Weight (kg)")
axes[0,0].set_xlabel("Height (cm)")


# Subplot 1,0 (bottom-left)
df_male['Height'].hist(bins='auto', edgecolor='black',ax=axes[1,0])
axes[1,0].set_ylabel('Counts per Bin', fontsize = 12);
axes[1,0].set_xlabel("Height (cm)")


## Plotting female data
# Subplot 0,1 (top-right)
sns.scatterplot(data=df_female, x='Height', y='Weight', ax=axes[0,1],
                color='orange')
axes[0,1].set_title('Female Height vs Weight', fontsize = 12)
axes[0,1].set_ylabel("Weight (kg)")


# Subplot 1,1 (bottom-right)
df_female['Height'].hist(bins='auto', edgecolor='black',ax=axes[1,1],
                         color='orange')
axes[1,1].set_ylabel('Counts per Bin', fontsize = 12);
axes[1,1].set_xlabel("Height (cm)")


## Figure title and fixing overlap
fig.suptitle("Example 2D Subplots comparing Males vs Females ");
fig.tight_layout();
```
