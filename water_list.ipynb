```python
water_list= [150, 200, 50, 300, 100, 450, 80]

def The_Evaluator(numberone_liters):
  if numberone_liters > 250 :
    return "high"

  elif numberone_liters <= 250 :
    return "Normal"

def The_Extremes(water_list):
  print (f" The lowest consumption was : {min(water_list)}")
  print (f" The hightest consumption was : {max(water_list)}")


def The_Manager(water_list):
 for day in water_list:
  result = The_Evaluator(day) # بنسأل المساعد: شو رأيك بهاد اليوم؟
  print(f"Day status: {result}") # بنطبع الجواب (High أو Normal)


The_Extremes(water_list)
The_Manager(water_list)

```
