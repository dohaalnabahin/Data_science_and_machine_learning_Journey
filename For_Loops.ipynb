```python
 #define the list to be iterated on
X = [1, 2, 3, 4, 5,]
# write a for loop
for i in X:
  double_number = i*2
  print(double_number)
print(X)
```

```python
# create an empty list to store the new values
X_double = []   # create list
for number in X:
  double_number = number*2 #transformation
  X_double.append(double_number) #append
  # print the new list
print(X_double)
```

```python
celsius = [0, 10, 20, 30]
fahrenheit =[]
for i in celsius :
  resualt = i * 1.8 + 32
  fahrenheit.append(resualt)
print(fahrenheit)
```

```python
#example of Looping through a List
temps_f = [ 65, 60, 55, 72, 67, 84]
temps_c = []
for f in temps_f :
  Celsius =  round((f - 32) * (5/9) ,2)
  temps_c.append(Celsius)
print(temps_c)


```

```python
# Example of Looping through a Dictionary

# Run this cell to create a dictionary with room type and standard rate
standard_dict = {'Basic': 100,
              'Deluxe': 125,
            'Balcony' : 200,
              'Luxury' : 250}
for room_type,price in standard_dict.items() :
  print(f'{room_type} rooms are available for only ${price} a night.')
```

```python
# For Loops to Append Items to a Dictionary
standard_dict = {'Basic': 100,'Deluxe': 125,'Balcony' : 200,'Luxury' : 250}
holiday_dict= {}
for room_type,price in standard_dict.items() :
  resulat =  price * 1.3
  holiday_dict[room_type] = resulat
print(holiday_dict)



```
