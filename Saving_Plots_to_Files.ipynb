```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Keep in mind that the filename will likely be different on your computer (especially later in course)
filename = '/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week03/linearRegPredicted.csv'
df = pd.read_csv(filename)
df.head()
```

```python
feature = df['feature']
actual = df['actual']
predicted = df['predicted']
```

```python
# The legend will be cutoff when saved
fig, axes = plt.subplots();
axes.plot(feature, predicted, c = 'r', label = 'Prediction')
axes.scatter(feature, actual, c= 'k', label = 'Actual')
axes.legend(loc=(1.02,0))

fig.savefig('legendcutoff.png', dpi = 300)
```

```python
# Save image properly
fig, axes = plt.subplots();
axes.plot(feature, predicted, c = 'r', label = 'Prediction')
axes.scatter(feature, actual, c = 'k', label = 'Actual')
axes.legend(loc=(1.02,0))
fig.tight_layout()
fig.savefig('SavedPlot.png', dpi = 300)
```
