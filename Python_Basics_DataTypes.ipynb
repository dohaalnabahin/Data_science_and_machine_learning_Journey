# Dictionaries
- A group of key-value pairs
- Dictionary elements are indexed by unique keys which are used to access values
-  في بايثون يشبه القاموس الحقيقي أو "دفتر العناوين". بدلاً من الوصول للعناصر عبر الأرقام ($0, 1, 2$), نصل إليها عبر أسماء (Unique Keys)

```python
empty_dict = {}
new_person = {'name': 'John', 'age': 38, 'weight': 160.2, 'has_glasses': False}
#new_person: هو اسم المتغير (Variable Name)
#{ }: الأقواس المتعرجة هي التي تُخبر بايثون أن هذا "قاموس" وليس قائمة

#'name' , 'age' , 'weight' , 'has_glasses':
#تُسمى Keys (مفاتيح).
#وظيفتها: تعمل كعنوان للوصول للمعلومة.
#يجب أن تكون فريدة (Unique) ولا تتكرر

#'John' , 38 , 160.2 , False:
#تُسمى Values (قيم).
#هي المعلومات الفعلية المخزنة داخل العناوين.
#لاحظي أن القيم متنوعة الأنواع (String, Integer, Float, Boolean)

# updates if the key exists
new_person['name'] = 'Jack'

# adds a key-value pair if the key doesn't exist
new_person['hobbies'] = ['climbing', 'coding']

print(new_person)
# output: {'name': 'Jack', 'age': 38, 'weight': 160.2, 'has_glasses': False, 'hobbies': ['climbing', 'coding']}

# removes the specified key and returns the value
w = new_person.pop('weight')
print(w)
# output: 160.2

print(new_person)
# output: {'name': 'Jack', 'age': 38, 'has_glasses': False, 'hobbies': ['climbing', 'coding']}
```

# String Interpolation

مكننا أيضاً إدخال متغيرات في سلاسلنا النصية، وهو ما يُعرف باسم استيفاء السلاسل النصية
## F-Strings (Literal String Interpolation)
لإنشاء سلسلة F، ضع الحرف f قبل علامة الاقتباس الافتتاحية مباشرةً. ثم ضع أي متغيرات داخل السلسلة بين قوسين معقوفين

## string.format()
استخدامها، اكتب النص كاملاً، واستبدل أي كلمات تستمد قيمها من متغيرات بـ {}. ثم استدعِ دالة التنسيق على النص، مع تمرير الوسائط بالترتيب الذي يجب أن تملأ به الفراغات بين {}

**Read from left to right**: يعني بايثون يقرأ النص من اليسار لليمين، وأول {} يراها يضع فيها أول قيمة موجودة داخل format().

**Same number of sets:** يعني لو وضعتِ {} مرتين في النص، يجب أن تمرري كلمتين (متغيرين) داخل format. لو كان هناك نقص أو زيادة، سيحدث خطأ

## %-formatting
بدلاً من الأقواس المعقوفة، يُستخدم رمز النسبة المئوية (%) للإشارة إلى عنصر نائب، %s للنص و%d للرقم. بعد النص، يفصل رمز النسبة المئوية (%) النص المراد دمجه عن القيم المراد إدراجها فيه،


```python
#F-Strings (Literal String Interpolation)
first_name = "Zen"
last_name = "Coder"
age = 27
print(f"My name is {first_name} {last_name} and I am {age} years old.")

#************************************

#string.format()
first_name = "Zen"
last_name = "Coder"
age = 27
# Correct order
print("My name is {} {} and I am {} years old.".format(first_name, last_name, age))
# output: My name is Zen Coder and I am 27 years old.
# Wrong order (The method follows the order of arguments)
print("My name is {} {} and I am {} years old.".format(age, first_name, last_name))
# output: My name is 27 Zen and I am Coder years old.

#************************************

#%-formatting
hw = "Hello %s" % "world"      # %s is a placeholder for a string
py = "I love Python %d" % 3    # %d is a placeholder for a number (decimal)

print(hw, py)
# output: Hello world I love Python 3

name = "Zen"
age = 27
print("My name is %s and I'm %d" % (name, age))
# output: My name is Zen and I'm 27
```

#Built-In String Methods
الطرق الأخرى التي يمكننا تطبيقها على سلسلة نصية


```python
x = "hello world"

# Capitalizes the first letter of every word
print(x.title())

# output: "Hello World"
```

```python
# 1. upper(): تحول جميع الحروف إلى كبيرة
print("hello world".upper())
# output: HELLO WORLD

# 2. lower(): تحول جميع الحروف إلى صغيرة
print("Hello World".lower())
# output: hello world

# 3. count(substring): تحسب كم مرة تكرر حرف أو كلمة معينة
print("hello world".count("l"))
# output: 3 (حرف الـ l تكرر 3 مرات)

# 4. split(char): تقسم النص إلى قائمة (List) عند علامة معينة
print("hello world".split(" "))
# output: ['hello', 'world'] (قسمها عند المسافة)
sentence = "Python is amazing"
words = sentence.split(" ")
print(words)
# الناتج: ['Python', 'is', 'amazing']


users = "Doha,Sara,Noor"
users_list = users.split(",")
print(users_list)
# الناتج: ['Doha', 'Sara', 'Noor']

# split cnvert from string to list with "" , ","

# 5. find(substring): تعطيكِ مكان (Index) أول حرف من الكلمة التي تبحثين عنها
print("hello world".find("world"))
# output: 6 (كلمة world تبدأ من الاندكس 6)

# 6. isalnum(): هل النص يتكون من حروف وأرقام فقط (بدون مسافات أو رموز)؟
text1 = "Doha2026"
print(text1.isalnum()) # output: True

text2 = "Doha 2026!"
print(text2.isalnum()) # output: False (بسبب المسافة وعلامة التعجب)

# 7. isalpha(): هل النص حروف فقط؟
print("Python".isalpha()) # output: True

# 8. isdigit(): هل النص أرقام فقط؟
print("123".isdigit())    # output: True




```
