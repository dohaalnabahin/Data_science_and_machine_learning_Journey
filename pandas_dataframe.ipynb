# الجانب العملي: مقدمة في Pandas DataFrames

```python


from re import T
import pandas as pd
import numpy as np

# إنشاء 3 قوائم (Names, Prices, Quantities) تمثل بيانات مشروع الـ Bake Sale.
names_list = ['brownie','cookie','cake', 'cupcake']
prices_list = [2.25,1.25,9.5, 3.5]
quantities_sold_list = [17, 40, 1, 10]

#تحويل القوائم إلى NumPy Arrays لتجربة الحسابات العنصرية (Element-wise calculations).
prices_array = np.array(prices_list)
quantities_sold_list_array = np.array(quantities_sold_list)

len(prices_list)  # list have length
len(prices_array) # array have list
prices_array.shape  # just array have shape

# Arrays perform operations elementwise
prices_array * quantities_sold_list_array   # بس في المصفوفة بنقدر نضرب لانها تدعم element-wise calculations. ام الليست لايمكن الضرب فيها


#إنشاء DataFrame من قاموس (Dictionary) يجمع القوائم الثلاث.
#df.shape
df = pd.DataFrame({'Name':names_list,
                   'Price':prices_list,
                   "Quantity Sold": quantities_sold_list})


#استخدام الاختصارات df.columns و df.values و df.index لفحص مكونات الجدول
df.columns
df.index
df.values

# تغيير الـ Index ليصبح عمود 'Name' بدلاً من الأرقام الافتراضية.
df.set_index("Name") # هنا لا يقوم بالتغيير على الجدول الاصلي

df = df.set_index("Name") # هنا عند لبحفظ في متغير يقوم بالتغيير في الجدول الاصلي
# df.set_index("Name" , inplace=True)   # هنا برضو طريقة ثانية للتغير على الجدول الاصلي
df
```

#الجانب العملي: العمل مع DataFrames (Slicing & Ops)

```python
# استخراج عمود واحد باستخدام الأقواس المفردة [] للحصول على Series.
#sliced.shape
sliced = df['Price']

#استخراج عمود واحد باستخدام الأقواس المزدوجة [[]] للحصول على DataFrame.
df_slice = df[['Price']]

#تقطيع عدة أعمدة معاً (Multiple Columns) وتخزين أسمائها في متغير.
cols_to_slice = ['Quantity Sold', 'Price']

#استخدام .shape للمقارنة بين أبعاد الـ DataFrame والـ Series.
df.shape
sliced.shape

 #إنشاء عمود 'Sales' جديد عبر ضرب عمودي السعر والكمية (Element-wise)
df['Sales'] = df['Price'] * df["Quantity Sold"] # هيك انشانا عمود وحفظناه في الجدول

#حساب مجموع المبيعات الكلي باستخدام دالة .sum().
df['Sales'].sum()

#إنشاء Series جديدة للمصاريف (Expenses) باستخدام قاموس مع تسميتها (name)
expenses = pd.Series({'brownie':0.25, 'cookie':0.5, 'cake':5, 'cupcake': 0.75},
                     name='Expenses')

#دمج (Concatenate) عمود المصاريف مع الجدول الأصلي باستخدام pd.concat مع تحديد axis=1.
df = pd.concat([df, expenses], axis=1)  # اذا كانت صفر يعني فوق بعض اذا كانت واحد يعني جنب بعض

# df
#حساب الربح لكل قطعة (Profit Per Item) والربح الإجمالي (Profits) وإضافتهما كأعمدة جديدة.
df = df.loc[:, ~df.columns.duplicated()]
df['Profit Per Item'] = df['Price'] - df['Expenses']  # calculate Profit Per Item
df["Profits"] = df['Profit Per Item'] * df['Quantity Sold']  # calculate profit
df
total_profit = df['Profits'].sum()
total_profit




```
