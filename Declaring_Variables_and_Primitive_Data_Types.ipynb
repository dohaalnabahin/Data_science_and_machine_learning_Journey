```python
# Does .upper() return a value or change a value in place?
message = 'hello'
caps = message.upper()
print(message)
print(caps)
```

```python
# How to overwrite an immutable object
message = 'hello'
mdoha = message.upper()
print(mdoha)
```

```python
message = "hello"

# هنا بايثون صنعت كلمة "HELLO" في مكان جديد في الذاكرة
# لكن message لسه بتأشر على "hello" الصغيرة!
message.upper()
print(message) # رح تطبع: hello (لأنها immutable)

# عشان تغيريها فعلياً، لازم تعملي Overwrite:
message = message.upper()
print(message) # رح تطبع: HELLO
```

```python
# Use an f-string to insert variable values
regular_price = 300
sale = .2
sale_price = (1-sale)*regular_price
statement = f'The orignal cost was ${regular_price}, but I purchased it on sale for ${sale_price}.'
print(statement)
```

```python
# Create a list object declared as the variable grocery
grocery = ['eggs', 'milk']
print(grocery)
# add "cheese" to our list. Does this change the object in place?
grocery.append('cheese')
print(grocery)
```

```python
# Define the variables
regular_price = 300
sale = .2
sale_price = (1-sale)*regular_price
# Check the data type of regular_price
type(regular_price)
# Check the data type of sale
type(sale)
type(sale_price)
```

```python
# Declare a whole number value (it will default to data type int)
amount = 6
type(amount)
# Recast the amount variable as a float
amount = int(amount)
type(amount)
```

```python
# Declare a decimal value (it will default to data type float)
value = 23.85
our_values = int(value)
type(our_values)
type(value)
print(our_values)
```

```python
# Declare a decimal value (it will default to data type float)
value = 23.85
our_values = round(value)
print(our_values)
```
