Countdown - Create a function that accepts a number as an input. Return a new list that counts down by one, from the number (as the 0th element) down to 0 (as the last element).


Example: countdown(5) should return the list: [5,4,3,2,1,0]



```python
def countdown(num):
    return list(range(num, -1, -1))
print(countdown(5))
```

This Length, That Value (Optional) - Write a function that accepts two integers as parameters: size and value. The function should create and return a list whose length is equal to the given size, and whose values are all the given value.


Example: length_and_value(4,7) should return [7,7,7,7]


Example: length_and_value(6,2) should return [2,2,2,2,2,2]

```python
def length_and_value(size, value):
    new_list = []
    for i in range(size):
        new_list.append(value)
    return new_list
print(length_and_value(4, 7))
```

Required: Write a function that takes a list of 3 numbers:

If the last number in the list is greater than 100, print: "Big ending!".

In any case, increment the value of the first number in the list by 10.

Finally, return the entire list.

```python
def Conditional_Increment(My_list):
    if My_list[2] > 100:
        print("Big ending!")

    My_list[0] = My_list[0] + 10

    return My_list

cool_numbers = [5, 50, 150]
result = Conditional_Increment(cool_numbers)
print(result)

```

If the total score is greater than 200, print: "Legendary Player!"

If the second number in the list is 100 (meaning they scored a perfect score), print: "Perfect Game 2!"

Finally, return the total score.

```python
def The_Pro_Player(My_list):
    total = My_list[0] + My_list[1] + My_list[2]

    if total > 200:
        print("Legendary Player!")

    if My_list[1] == 100:
        print("Perfect Game 2!")

    return total


cool_numbers = [50, 100, 70]
result = The_Pro_Player(cool_numbers)
print("Total Score:", result)

```

Simple Currency Converter Challenge

Required: Write a function that takes an amount and a currency name, and converts it to Jordanian Dinars based on the selected currency:

If the currency is USD (US Dollar), multiply the amount by 0.70.

If the currency is EUR (Euro), multiply the amount by 0.80.

If it's any other currency, print "Currency not supported" and return the original amount.

```python
def convert(amount, currency):
    if currency == "USD":
        result = amount * 0.70
        print("Converted to USD:", result)
        return result

    elif currency == "EUR":
        result = amount * 0.80
        print("Converted to EUR:", result)
        return result

    else:
        print("Currency not supported for:", amount)
        return amount

final_total = convert(100, "USD")
print("Final Result in JD:", final_total)
```
