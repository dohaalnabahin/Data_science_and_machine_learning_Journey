```python
# 1. المساعد الأول (لازم يكون مكتوب فوق)
def to_percent(mark):
    return str(mark) + "%"

# 2. المساعد الثاني (تأكدي من كتابة هذا الجزء)
def grade_summary(marks_list):
    top = max(marks_list)
    lowest = min(marks_list)
    print(f"Top Grade: {top}")
    print(f"Lowest Grade: {lowest}")

# 3. الدالة الرئيسية (المدير)
def process_grades(all_marks):
    print("--- Students Percentages ---")
    for m in all_marks:
        print(to_percent(m)) # مناداة المساعد 1

    print("--- Summary ---")
    grade_summary(all_marks) # مناداة المساعد 2

# 4. تشغيل الكود
my_class = [85, 92, 78, 95, 60]
process_grades(my_class)
```
