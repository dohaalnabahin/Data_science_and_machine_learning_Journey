```python
# Import needed libraries
import pandas as pd

# Mount google drive
from google.colab import drive
drive.mount('/content/drive')

# This filepath will work if you saved your file in the recommend folder structure
fname = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/Our Bake Sale [Loading Data Vers] - Day 4 Sales.csv"
df = pd.read_csv(fname)
df.head()


# Review: slicing 1 column
df["Item"] #series

# Review: slicing more than 1 column
df[["Item", "Student"]]   #DataFrame

# Slice the row with cake at index 2
df.loc[2]  # series
df.loc[[2]]   # dataframe

df = df.set_index("Item")
# df.head()

#  # Slice out the row with cake as the index
df.loc[['cake']]

# Slice multiple rows using a list
df.loc[['cake','brownie']]


# Slice the muffin row and Price column
print(df.loc["muffin",'Price'])

# Slice Price and Student for muffin
df.loc["muffin", ["Price","Student"]]

#What if we wanted the Price, Expense, and Baked By columns for muffins, scones, and banana bread?
df.loc[ ['muffin','scone','banana bread'], ['Price','Expense','Student']]


# Save lists of rows
rows_to_slice =  ['muffin','scone','banana bread']

# Save list of columns
cols_to_slice =  ['Price','Expense','Student']

# Use variables of lists to slice
df.loc[rows_to_slice, cols_to_slice]

# Selet all rows for the Price Column
df.loc[:, 'Price']


# Slice all rows for mulitple columns
df.loc[:, ["Price","Expense"]]

# Locate the Price of a torte and change the value
df.loc['torte',"Price"] = 11.5

df.loc[:, ["Price","Expense"]]

# Inspect torte
df.loc['torte']

# Check the price of apple pie
price = df.loc['apple pie','Price']
print(price)

# is the price greater than 5?
price > 5

# Is the price greater than 5? (for all rows)
df['Price'] > 5

# Apply a boolean index to filter our data
df.loc[ df['Price'] > 5 ]

# Save the boolean index as a separate variable
filter_price = df["Price"] > 5
filter_price

# Apply the filter to create a dataframe of selected values using square brackets
df_filtered = df[filter_price]
df_filtered


# Apply the filter to create a dataframe of selected values using .loc
df_filtered = df.loc[filter_price]
df_filtered

# Get the sum of True values for the filter
filter_price.sum()


# Get the length of the filtered datarame
len(df_filtered)

# Create a filter for Packaging values that contain "individual"
filter_individual = df['Packaging'].str.contains('Individual')
# Apply the filter
df.loc[filter_individual]

# Use the inversion operator to get all values that do NOT meet the filter conditions
df[~filter_individual]

# Create a filter for any index value that contains "pie"
filter_contains_pie = df.index.str.contains('pie')
# Apply the filter
df[filter_contains_pie]


# Create a filter for any index value that contains "pie"
filter_contains_pie = df.index.str.contains(' pie')
# Apply the filter
df[filter_contains_pie]

# Create a filter for Jade
filter_jade = df['Student']=='Jade'
filter_jade

# Create a filter for Price
filter_price = df['Price'] > 5
filter_price

# Combine the filters
filters_combined = filter_jade & filter_price
filters_combined


# Apply the combined filter
df.loc[filters_combined]

# Create a new filter for less than 2
filter_price_less = df['Price'] < 2
# Combine the filters
df.loc[ filter_price_less & filter_jade]


# Alternate method for combing filters without saving them first
df.loc[(df['Student']=='Jade') & (df['Price'] > 5)]


# List of students
students = ["Tom","Anne",'Martin', "Jose",'Sally']

# هذا الكود هو "الفحص". بايثون بيمسك عمود الطلاب اسم اسم، وبيسأل نفسه: "هل هذا الاسم موجود داخل قائمة students اللي عندي؟"
# Creating a filter for any student in the list
df['Student'].isin(students)


#"يا بايثون، اعرض لي صفوف الطلاب اللي طلع فحصهم True (يعني اللي أساميهم موجودة في القائمة)"
# Apply the filter
df[df['Student'].isin(students)]






# # another example
# # This filepath will work if you saved your file in the recommend folder structure
# fname = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/Titanic-Dataset.csv"
# df = pd.read_csv(fname)
# df.head()
# # استخراج الصف رقم 2
# row_2 = df.loc[2]
# print(row_2)
# # تعيين الاسم كفهرس
# df_names = df.set_index("Name")
# # الآن لنبحث عن راكب معين بالاسم (بدل cake)
# passenger_info = df_names.loc['Heikkinen, Miss. Laina']
# print(passenger_info)
# # row_2 = df.loc[2]



```
